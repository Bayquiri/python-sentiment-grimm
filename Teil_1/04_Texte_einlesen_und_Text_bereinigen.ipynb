# Texte einlesen und bereinigen

Wenn wir uns innerhalb der Geisteswissenschaften mit Python bewegen, ist die Arbeit mit Texten sehr relevant. Mit Python können wir Texte speichern, verändern und sogar ganze Textdateien einlesen und auswerten lassen. 

Beginnen wir mit der Bearbeitung von Texten. Wie wir bereits wissen, können Texte bzw. Strings in Variablen gespeichert werden. Nehmen wir bspw. den Titel des Werks "Die Sterntaler":


````python
titel = "Die Sterntaler"
print(titel)
````

Erwarteter Output: `Die Sterntaler`

Wir können den Text auch verändern und normalisieren.

---

### Hinweis 

Normalisieren bedeutet in diesem Kontext, dass ein Text vereinheitlicht wird, damit das Programm gleiche Inhalte als solche identifizieren kann. Ein Schritt der Normalisierung ist die sog. "Case Conversion", bei der der gesamte Text in Kleinbuchstaben überführt wird. Denn wie wir bereits gelernt haben, grenzt Python nach Groß- und Kleinschreibung ab ("Märchen" und "märchen" wären zwei unterschiedliche Werte).

---

Die Kleinschreibung können wir zunächst mit ".lower()" erreichen:


````python
print(titel.lower())
````

Erwarteter Output: `die sterntaler`

Aber Achtung: die kleingeschriebene Variante ist nicht in der Variablen gespeichert. Sie wurde nur einmalig ausgegeben.

## Text einlesen

Nachdem wir den Titel nun für Python normalisiert haben, wollen wir den Text des Märchens einlesen. Aber wie machen wir das? Wir müssen zum Glück nicht den gesamten Märchentext in einen String schreiben und speichern. Denn wir können ganze (Text-)Dateien einlesen, in einer Variablen speichern und dann später verändern bzw. auswerten. 

Hierzu nutzen wir in einem ersten Schritt die Funktion `with open('Dateipfad zum Text', 'r', encoding='utf-8') as datei:` und in einem zweiten die konkrete Anweisung zum Lesen der Datei mit `sterntaler = datei.read()`.

*Hinweis:* Das `'r'` steht für "read", also, dass die Datei gelesen wird aber die Originaldatei nicht von uns geändert oder überschrieben wird. Das Encoding nach utf-8 hilft, dass Sonderzeichen bspw. richtig angezeigt werden.

Wir nehmen das Märchen "Die Sternthaler":


````python
with open('maerchen_texte/153_Die_Sternthaler.txt', 'r', encoding='utf-8') as datei:
    sterntaler = datei.read()
````

Nun ist der gesamte Text in der Variablen *sterntaler* gespeichert. Wir können uns anzeigen lassen, wie lang der Text ist, also die Anzahl der Zeichen ausgeben lassen:


````python
print(len(sterntaler))
````

Erwarteter Output: `1568`

Wir haben nun die Information erhalten, dass das Märchen "Die Sternthaler" 1568 Zeichen enthält. Gemeint sind damit keine Wörter, sondern tatsächlich einzelne Zeichen: Buchstaben, Satzzeichen, womöglich auch Zahlen.

## Text bereinigen

Um valide mit dem Text arbeiten zu können, müssen wir auch im Text für Einheit und Vergleichbarkeit sorgen. So, wie wir den Titel des Märchens normalisiert haben, können bzw. müssen wir das auch für den gesamten Text tun. 

Damit Wörter vergleichbar sind und Satzzeichen nicht stören, können wir mit einfachen Methoden alles klein schreiben, die Punkte und Kommata durch "Nichts" ersetzen und dann den Text in eine Liste von Wörtern zerlegen.

Für die Kleinschreibung aller Wörter im Text nutzen wir die eingebaute Funktion lower():


````python
sterntaler_klein = sterntaler.lower()
````

Auch Satzzeichen brauchen wir für unser Vorhaben nicht. Diese können wir ebenfalls mit einer von Python vorgegebenen Funktion direkt entfernen: `replace()`

Wichtig ist: Wir bestimmen mit der Funktion, worauf sie angewendet werden soll und was durch etwas anderes ersetzt wird. Hier sagen wir Python also: Nimm den kleingeschriebenen Text und ersetze den Punkt `"."` mit `""` Nichts (Anführungszeichen ohne Inhalt).


````python
bereinigter_text = sterntaler_klein.replace(".", "").replace(",", "")
````

Nun zerlegen wir den gesamten Text in einzelne Wörter und speichern diese in einer Liste. Dafür wenden wir die eingebaute Funktion `split()` an und lassen uns zur Kontrolle die ersten 10 Listenelemente ausgeben:

`split()` zerlegt eine Zeichenkette in einzelne Elemente und trennt am Leerzeichen.

*Hinweis:* Man kann selbst bestimmen, an welchem Zeichen getrennt wird. Mehr Informationen: https://www.w3schools.com/python/ref_string_split.asp


````python
wort_liste = bereinigter_text.split()
# gib die ersten 10 Wörter aus:
print(wort_liste[:10])
````

## Wörter zählen

Wenn wir nun wissen wollen, aus wie vielen Wörtern (nicht Zeichen!) unser Text bzw. unsere Liste besteht, wäre es sehr müßig, die einzelnen Wörter selbst nachzuzählen. Glücklicherweise können wir in Python dafür sorgen, dass solche Vorgänge automatisiert stattfinden: Mit Bibliotheken.

### Bibliotheken

Wörter sind nützlich, aber noch nützlicher sind die Sätze und Geschichten, die wir mit ihnen bauen. Ähnlich verhält es sich mit Python: Während viele leistungsstarke, allgemeine Werkzeuge in Python eingebaut sind, leben spezialisierte Werkzeuge in Bibliotheken, die bei Bedarf aufgerufen werden können.

Eine Python-Bibliothek ist eine Sammlung von Dateien (genannt Module), die Funktionen enthält, die Sie in Ihren Programmen verwenden können. Einige beliebte Bibliotheken sind:

- **Pandas** - Werkzeug zur tabellarischen Datenanalyse
- **Matplotlib** - Werkzeuge zur Datenvisualisierung
- **BeautifulSoup** - für das Parsen von HTML- und XML-Dokumenten
- **Requests** - für HTTP-Anfragen (z. B. für Web-Scraping, Verwendung von APIs)
- **Scikit-learn** - Werkzeuge zum maschinellen Lernen
- **NumPy** - Werkzeuge für numerische Berechnungen

#### Import von Bibliotheken

Verwenden Sie `import`, um eine Bibliothek in den Speicher eines Programms zu laden:


````python
import string

print(f'The lower ascii letters are {string.ascii_lowercase}')
print(string.capwords('capitalise this sentence please.'))
````

Erwarteter Output:
```
The lower ascii letters are abcdefghijklmnopqrstuvwxyz
Capitalise This Sentence Please.
```

---

Um die Wörter in unserer Liste zu zählen, brauchen wir eine der Bibliotheken, die Python zur Verfügung stellt:


````python
from collections import Counter

# Der Counter erledigt das Zählen für uns
haeufigkeiten = Counter(wort_liste)
````

Wir können uns über die Funktion `most_common()` die häufigsten gezählten Wörter anzeigen lassen:


````python
# Die 10 häufigsten Wörter ausgeben
print(haeufigkeiten.most_common(10))
````
