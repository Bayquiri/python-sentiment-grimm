# Listen


Eine der beliebtesten Arten der Datensammlung in Python ist die Liste. Sie haben zwei wichtige Eigenschaften:

- Sie sind veränderbar, d.h. sie können nach ihrer Erstellung verändert werden
- Sie sind heterogen, d.h. sie können Werte vieler verschiedener Typen speichern (anders als in anderen Programmiersprachen)

Um eine neue Liste zu erstellen, können Werte in eckige Klammern und durch Kommata getrennt geschrieben werden. Handelt es sich um Zeichenketten / Strings, müssen die Werte innerhalb der Liste in einfache oder doppelte Anführungszeichen gesetzt werden:


````python
titel_märchen = ['Der Froschkönig', 'Rotkäppchen', 'Hänsel und Gretel', 'Schneewittchen']
titel_märchen
````

Wir können uns auch dank der Funktion `len()` anzeigen lassen, wie viele Werte in der Liste stehen:


````python
len(titel_märchen)
````

## Benutze einen Index und Slice, um Elemente aus der Liste anzuzeigen

Auf die gleiche Weise, wie wir Indexnummern für Strings genutzt haben, können wir damit auf Elemente und Stücke in einer Liste verweisen:


````python
print('Erstes Märchen der Liste:', titel_märchen[0])
print('Die drei ersten Märchen:', titel_märchen[0:3])
````

## Listenwerte mit ihrem Index neu zuordnen

Verwenden Sie einen Indexwert zusammen mit ihrer Listenvariablen, um einen Wert in der Liste zu ersetzen:


````python
print('Liste war:', titel_märchen)
titel_märchen[0] = 'Rumpelstielzchen'
print('Liste jetzt:', titel_märchen)
````

---

### Hinweis: Zeichenketten sind unveränderlich

Im Gegensatz zu Listen können wir die Zeichen in einer Zeichenkette nicht über ihren Indexwert ändern. Mit anderen Worten: Zeichenketten sind unveränderlich, sie können nach ihrer Erstellung nicht mehr geändert werden. Denn: Python betrachtet die Zeichenkette als einen einzelnen Wert bestehend aus verschiedenen Teilen und nicht als eine Sammlung von Werten, so wie es bei den Listen der Fall ist.

---

## Listen können unterschiedliche Datentypen enthalten

Eine einzelne Liste kann Zahlen, Zeichenketten und alles andere enthalten (auch andere Listen!). Wenn es sich um eine Liste innerhalb einer Liste handelt, können Sie weiterhin die Notation in eckigen Klammern verwenden, um auf bestimmte Elemente zu verweisen.


````python
mixed_list = ['string', 3.2, [10, 20, 30]]
print('Erstes Item in zweiter Liste:', mixed_list[2][0])
````

## Erweitern einer Liste

Verwenden Sie den Zusatz `append` (=anhängen), um der Liste weitere Elemente hinzuzufügen, die dann am Listenende erscheinen. Bei "append" handelt es sich um eine Methode. Methoden können nicht für sich allein stehen, sondern brauchen immer ein Objekt, auf das sie sich beziehen – im Gegensatz zu Funktionen, die eigene Codeblöcke einleiten. Das zeigt sich auch in der Schreibweise: Während Funktionen wie `print()` oder `type()` vorangestellt werden und die Klammer eröffnen, werden Methoden wie `.append` an das Objekt, das sie betreffen, angehängt.

Die Methode append wird von der Syntax mit `.append()` verwendet bzw. aufgerufen:


````python
print('Liste war:', titel_märchen)
titel_märchen.append('Die Bremer Stadtmusikanten')
print('Liste jetzt:', titel_märchen)
````

Durch `.append` konnten wir unserer Liste nun einen weiteren Märchentitel, Die Bremer Stadtmusikanten, zufügen.

## Elemente aus der Liste entfernen

Verwenden Sie `del`, um Elemente aus einer Liste vollständig zu entfernen.

`del list_name[index]` entfernt ein Element aus einer Liste und kürzt die Liste. Im Gegensatz zu `.append()` ist `del` keine Methode, sondern eine „Anweisung" in Python. Im folgenden Beispiel führt `del` eine „in-place"-Operation an unserer Liste durch. Das bedeutet, dass die Variable `titel_märchen` neu zugewiesen wird, wenn Sie die `del`-Anweisung verwenden, ohne dass Sie einen Zuweisungsoperator (z.B. `titel_märchen = ...`) verwenden müssen.


````python
titel_märchen = ['Rumpelstielzchen', 'Rotkäppchen', 'Hänsel und Gretel', 'Schneewittchen', 'Die Bremer Stadtmusikanten']
print('Liste vorher:', titel_märchen)
del titel_märchen[4]
print('Liste nachher:', titel_märchen)
````

Durch `del` mit dem Index `[4]` konnten wir nun das fünfte Element aus der Liste der Märchentitel entfernen.

### Hinweise

**1. Funktion (Function)**
Eine Funktion ist ein unabhängiger Codeblock, der eine bestimmte Aufgabe ausführt. Sie kann Parameter entgegennehmen und einen Wert zurückgeben (oder auch nicht). Funktionen sind nicht an ein bestimmtes Objekt oder eine Klasse gebunden. Sie werden einfach über ihren Namen aufgerufen (`funktionsname()`).
Beispiele: `print()`, `type()`, `slice()`

**2. Methode (Method)**
Eine Methode ist im Grunde ähnlich wie eine Funktion, die aber an ein Objekt gebunden ist. Sie operiert typischerweise mit den Daten (Attributen) dieses Objekts. Methoden werden über die Punkt-Notation aufgerufen: `objekt.methode()`.
Beispiel: `.append()`

**3. Anweisung (Statement)**
Eine Anweisung ist eine einzelne, vollständige Codezeile, die vom Python-Interpreter ausgeführt wird und eine Aktion bewirkt. Ein Python-Programm besteht aus einer Folge von Anweisungen. Anweisungen sind die grundlegenden Bausteine des Programmcodes. Sie umfassen Dinge wie Zuweisungen (`=`) oder den Aufruf von Funktionen/Methoden.
Beispiel: `del`
